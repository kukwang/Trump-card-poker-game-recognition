```python
# make train.txt, test.txt
import glob
import os
import pickle
import xml.etree.ElementTree as ET
from os import listdir, getcwd
from os.path import join

# train.txt 생성 후 주석 위치 바꿔서 test.txt 생성
#dirs = '/content/darknet/data/train'
dirs = '/content/darknet/data/test'

def getImagesInDir(dir_path):
    image_list = []
    for filename in glob.glob(dir_path + '/*.png'):
        image_list.append(filename)
    for filename in glob.glob(dir_path + '/*.jpg'):
        image_list.append(filename)
    for filename in glob.glob(dir_path + '/*.JPG'):
        image_list.append(filename)
    return image_list

full_dir_path = dirs

image_paths = getImagesInDir(full_dir_path)
print(full_dir_path)
#list_file = open('/content/darknet/data/train' + '.txt', 'w')
list_file = open('/content/darknet/data/test' + '.txt', 'w')

for image_path in image_paths:
    list_file.write(image_path + '\n')
        
list_file.close()

print("Finished processing")
```
